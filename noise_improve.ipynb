#### Apply filters to reduce noise and improve the clarity of moving objects

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(image):
    plt.imshow(image)
    plt.show()
    return

def display_image_gray(image):
    plt.imshow(image,cmap='gray')
    plt.show()
    return
```

```python
def gaussian_blur(image):
    image = cv2.GaussianBlur(image,(5,5),0)
    return image
```

```python
def apply_filter(image):
    filtered = cv2.bilateralFilter(image,9,75,75)
    
    return filtered
```

```python
def sharpen(image):
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    image = cv2.filter2D(image, -1, kernel)
    return image
```

```python
def median_blur(image):
    image = cv2.medianBlur(image,5)
    return image
```

```python
def noise_removal(image):
    image = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
    return image
```

```python
def threshold(image):
    ret,thresh = cv2.threshold(image,117,220,0)
    return thresh
```

```python
def noise_pipline(image_path):
    image = cv2.imread(image_path)
    orginal_image = image.copy()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = gaussian_blur(image)
    image = apply_filter(image)
    image = sharpen(image)
    image = noise_removal(image)
    _, ax = plt.subplots(1,2,figsize=(10,5))
    ax[0].imshow(orginal_image)
    ax[0].set_title('Original')
    ax[0].axis('off')
    ax[1].imshow(image)
    ax[1].set_title('Enhanced')
    ax[1].axis('off')
    plt.show()
    
```

```python
image_path = "images/image_2.png"
noise_pipline(image_path)
```
